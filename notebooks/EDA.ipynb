```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv(r"D:\Movie_Rating_Prediction\notebooks\data\IMDb Movies India.csv",encoding="windows-1252")
```

```python
data.head()
```

```python
data.dropna(subset = ['Name','Year','Duration','Genre','Rating','Votes','Director'],inplace = True)
data['Duration'] = data['Duration'].str.replace('min', '').astype(int)
data['Year'] = data['Year'].str.strip('()').astype(int)
data['Votes'] = data['Votes'].str.replace(",","").astype(int)
```

```python
data.head()
```

```python
sns.set_style('darkgrid')
sns.set_palette('Dark2')
```

```python
sns.pairplot(data = data)
```

```python
sns.set_palette('gnuplot2_r')
sns.lineplot(data=data, x="Rating", y="Votes")
```

```python
sns.set_palette('YlOrRd_r')
sns.histplot(x="Rating",data = data)
```

```python
data.drop(['Name','Director','Genre','Actor 1','Actor 2','Actor 3'],inplace = True,axis = 1)
```

```python
sns.heatmap(data = data.corr(),annot = True,cmap = 'viridis')
```
